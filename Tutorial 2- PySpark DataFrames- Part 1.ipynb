```python
### Adding Columns in data frame
df_pyspark=df_pyspark.withColumn('double_median_income',df_pyspark['median_income']*2)
```

```python
df_pyspark.show(5)
```

```python
### Drop the columns
df_pyspark=df_pyspark.drop('double_median_income')
```

```python
df_pyspark.show(5)
```

```python
### Rename the columns
df_pyspark.withColumnRenamed('longitude','Longitude').show(4)
```
